```python
import scrapy
from scrapy.crawler import CrawlerProcess
from scrapy.selector import Selector
```

```python
class TyGiaVangSpider(scrapy.Spider):
    name = "TyGiaVang"
    start_urls=['https://www.24h.com.vn/gia-vang-hom-nay-c425.html',]
    
    def parse(self, response):
        sel = Selector(response)
        #ngay = sel.xpath('//span[@class="inputDate"]/text()').get()
        ngay= sel.css('span.inputDate::text').get()
        ress = response.css('div.tabBody > table > tbody >tr')
        for res in ress:
            loai = res.css('h2')[0]
            giamua = res.css('span.fixw')[0]
            giaban = res.css('span.fixw')[1]
            yield{"ngay": ngay, "loai": loai.extract(), "giamua": giamua.extract(), "giaban": giaban.extract()}
```

```python
process = CrawlerProcess()
process.crawl(TyGiaVangSpider)
process.start()
```
